```python
# Use matplot lib for visualization
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df_results = pd.read_csv('performance_results.csv')
df_results.head(2)
```

```python
plt.figure(figsize=(10, 6))

bar_plot = sns.barplot(x='Architecture', y='Average_Latency_ms', data=df_results, errorbar=None)

for container in bar_plot.containers:
    bar_plot.bar_label(container)
    
bar_fig = bar_plot.get_figure()
bar_fig.savefig('Avg_Latency_barplot.png',dpi=300, bbox_inches='tight')
plt.show()
```
