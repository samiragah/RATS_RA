```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1.   
print(" Creating performance data...")
data = {
    "Architecture": ["RATS", "Monolithic"],
    "Average_Latency_ms": [20257.756324, 4924.160218],  # ms 
    "Min_Latency_ms": [15000.0, 3000.0],  #   min latency   
    "Max_Latency_ms": [24277.076960, 9975.248575],      # ms 
    "Throughput_req_per_sec": [29, 30]      # req/sec 
}

df_results = pd.DataFrame(data)

# 2.    CSV
df_results.to_csv("performance_results.csv", index=False)
print(" Results saved to performance_results.csv")
print("\n Performance Results:")
print(df_results.to_string(index=False))

# 3.  
print("\n Creating professional visualizations...")

#   
plt.style.use('default')
sns.set_palette("mako")
colors = ['#FF6B6B', '#4ECDC4']  # colors

#  1: Average Latency
plt.figure(figsize=(12, 8))
ax1 = sns.barplot(x='Architecture', y='Average_Latency_ms', data=df_results, 
                 palette=colors, errorbar=None, width=0.6)

#     
for container in ax1.containers:
    ax1.bar_label(container, fmt='%.0f ms', padding=3, fontsize=11, fontweight='bold')

plt.title('Average Latency Comparison\nRATS vs Monolithic Architecture', 
          fontsize=12, fontweight='bold', pad=20)
plt.xlabel('Architecture', fontsize=12, fontweight='bold')
plt.ylabel('Average Latency (ms)', fontsize=12, fontweight='bold')
plt.grid(axis='y', alpha=0.3)

#     
latency_improvement = ((df_results['Average_Latency_ms'].iloc[1] - df_results['Average_Latency_ms'].iloc[0]) / 
                       df_results['Average_Latency_ms'].iloc[1]) * 100
plt.annotate(f'{latency_improvement:+.1f}% improvement', 
             xy=(0.5, 0.7), xycoords='axes fraction',
             ha='center', va='center', fontsize=12,
             bbox=dict(boxstyle="round,pad=0.3", fc="aqua", alpha=0.7))

plt.tight_layout()
plt.savefig('Avg_Latency_comparison.png', dpi=300, bbox_inches='tight')
print(" Average Latency chart saved as 'Avg_Latency_comparison.png'")
plt.show()

#  2: Throughput
plt.figure(figsize=(12, 8))
ax2 = sns.barplot(x='Architecture', y='Throughput_req_per_sec', data=df_results, 
                 palette=colors, errorbar=None, width=0.6)

for container in ax2.containers:
    ax2.bar_label(container, fmt='%.0f req/s', padding=3, fontsize=11, fontweight='bold')

plt.title('Throughput Comparison\nRATS vs Monolithic Architecture', 
          fontsize=12, fontweight='bold', pad=20)
plt.xlabel('Architecture', fontsize=12, fontweight='bold')
plt.ylabel('Throughput (requests/second)', fontsize=12, fontweight='bold')
plt.grid(axis='y', alpha=0.3)

throughput_improvement = ((df_results['Throughput_req_per_sec'].iloc[0] - df_results['Throughput_req_per_sec'].iloc[1]) / 
                         df_results['Throughput_req_per_sec'].iloc[1]) * 100
plt.annotate(f'{throughput_improvement:+.1f}% improvement', 
             xy=(0.5, 0.7), xycoords='axes fraction',
             ha='center', va='center', fontsize=12,
             bbox=dict(boxstyle="round,pad=0.3", fc="aqua", alpha=0.7))

plt.tight_layout()
plt.savefig('Throughput_comparison.png', dpi=300, bbox_inches='tight')
print(" Throughput chart saved as 'Throughput_comparison.png'")
plt.show()

#  3: Range Plot (Min, Avg, Max Latency)
plt.figure(figsize=(14, 8))

#     range plot
architectures = df_results['Architecture'].tolist()
avg_latency = df_results['Average_Latency_ms'].tolist()
min_latency = df_results['Min_Latency_ms'].tolist()
max_latency = df_results['Max_Latency_ms'].tolist()

x_pos = np.arange(len(architectures))

plt.bar(x_pos, avg_latency, color=colors, alpha=0.7, label='Average Latency', width=0.6)

#     min  max
for i, (min_val, max_val) in enumerate(zip(min_latency, max_latency)):
    plt.plot([i, i], [min_val, max_val], color='black', linewidth=2, marker='_', markersize=10)
    plt.plot(i, min_val, 'o', color='green', markersize=8, label='Min Latency' if i == 0 else "")
    plt.plot(i, max_val, 'o', color='red', markersize=8, label='Max Latency' if i == 0 else "")

plt.xticks(x_pos, architectures)
plt.title('Latency Range Comparison\n(Min, Average, Max Latency)', fontsize=12, fontweight='bold', pad=20)
plt.xlabel('Architecture', fontsize=12, fontweight='bold')
plt.ylabel('Latency (ms)', fontsize=12, fontweight='bold')
plt.legend()
plt.grid(axis='y', alpha=0.3)

#   
for i, (avg, min_val, max_val) in enumerate(zip(avg_latency, min_latency, max_latency)):
    plt.text(i, avg + 1000, f'{avg:.0f} ms', ha='center', va='bottom', fontweight='bold')
    plt.text(i, min_val - 2000, f'Min: {min_val:.0f} ms', ha='center', va='top', fontsize=9, color='green')
    plt.text(i, max_val + 1000, f'Max: {max_val:.0f} ms', ha='center', va='bottom', fontsize=9, color='red')

plt.tight_layout()
plt.savefig('Latency_Range_comparison.png', dpi=300, bbox_inches='tight')
print(" Latency Range chart saved as 'Latency_Range_comparison.png'")
plt.show()

# 4.   
print("\n" + "="*60)
print(" PERFORMANCE REPORT SUMMARY")
print("="*60)

print(f"\n  Architecture Comparison:")
print(f"   - RATS: {df_results['Average_Latency_ms'].iloc[0]:.0f} ms avg latency")
print(f"   - Monolithic: {df_results['Average_Latency_ms'].iloc[1]:.0f} ms avg latency")

print(f"\n Performance Metrics:")
print(f"   - Latency Improvement: {latency_improvement:+.1f}%")
print(f"   - Throughput Difference: {throughput_improvement:+.1f}%")
print(f"   - RATS is {df_results['Average_Latency_ms'].iloc[1]/df_results['Average_Latency_ms'].iloc[0]:.1f}x faster")

print(f"\n Files Created:")
print("   - performance_results.csv")
print("   - Avg_Latency_comparison.png")
print("   - Throughput_comparison.png")
print("   - Latency_Range_comparison.png")

print("\n Performance analysis completed successfully!")
```
