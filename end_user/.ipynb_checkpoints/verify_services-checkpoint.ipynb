```python
import httpx
import asyncio

async def test_services():
    services = {
        "Execution Core (8001)": "http://localhost:8001/generate-signal",
        "Intermediary (8000)": "http://localhost:8000/request-signal", 
        "Monolithic (8002)": "http://localhost:8002/request-signal-mono"
    }
    
    async with httpx.AsyncClient() as client:
        for name, url in services.items():
            try:
                response = await client.post(url, json={"symbol": "AAPL"}, timeout=5.0)
                print(f" {name}: {response.status_code} - {response.json()}")
            except Exception as e:
                print(f" {name}: ERROR - {str(e)}")

print("Testing service health...")
await test_services()
```
